```python
import pandas as pd
import numpy as np
```

```python
data_path = "/data1/morrisq/yuj13/DLBio/deepseed/data/ecoli_mpra_expr.csv"
data = pd.read_csv(data_path)
data['realB'] = data['realB'].str.lower()
seq_data = data.realB.to_numpy(dtype=str)
strength_data = data['expr'].to_numpy(dtype=np.float32)
```

```python
np.save('Ecoli_new_165_seq.npy', seq_data)
np.save('Ecoli_new_165_y.npy', strength_data)
```

```python
with open("Ecoli_new_blast_promoter50.fasta", "w") as fasta_file:
    for i, seq in enumerate(seq_data):
        fasta_file.write(f">t{i}\n{seq}\n")
```
